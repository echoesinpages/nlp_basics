```python
# f(w) = 3(w - 11)² + 17

class minimize_quadratic(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.w = torch.nn.Parameter(torch.tensor([0.0]))
    def forward(self):
        g = self.w
        h = (g - 11)
        i = h**2
        j = 3 * i
        f = j + 17
        return f

model = minimize_quadratic()

optimizer = torch.optim.SGD(model.parameters(), lr=0.05)
tol = 1e-5
epoch = 0

print("Starting optimization (-_-)\n")

while True:
    current_value = model()

    optimizer.zero_grad()
    current_value.backward()
    optimizer.step()

    new_value = model()

    if torch.abs(new_value - current_value) < tol:
        break

    epoch += 1
    print(f"Epoch {epoch:3d}   f(w) = {current_value.item():9.4f}    w = {model.w.item():8.4f}")

print("\n" + "─"*60)
print(f"Done in {epoch} steps!")
print(f"The minimum is at w = {model.w.item():.4f}")
print(f"Function value   = {model().item():.4f}")
print(f"(True answer was w = 11, f = 17)")
print("─"*60)
```
